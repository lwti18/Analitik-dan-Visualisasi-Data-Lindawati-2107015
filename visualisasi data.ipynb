```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
filelocation = 'D:\TUGAS LINDA\MATERI SEMESTER 5\ANALITIK DAN VISUALISASI DATA\covid19.csv'
df = pd.read_csv(filelocation)
```

```python
df.head()
```

```python
df['Country/Region'] == 'Albania'
```

```python
Albania = df[df['Country/Region'] == 'Albania']
Albania
```

```python
kasus = Albania[Albania['Confirmed']>0]
kasus
```

```python
plt.plot(kasus['Date'], kasus['Confirmed'])
plt.show()
```

```python
kasusjuni = kasus[kasus.Date >= '6/1/20']
kasusjuni
```

```python
kasusmei = kasus[kasus.Date >= '5/1/20']
kasusmei
```

```python
plt.plot(kasusmei.Date, kasusmei.Confirmed)
plt.plot(kasusmei.Date, kasusmei.Recovered)
plt.legend(['Jumlah kasus', 'Sembuh'])
plt.xticks(rotation = 'vertical')
plt.show()
```
